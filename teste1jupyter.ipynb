```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('C:/Users/rafae/Desktop/songs/SpotifyMostStreamedSongs.csv')
df
```

```python
filter1 = df.loc[df['released_year'] == 2023]
filter1
```

```python
df['streams'] = pd.to_numeric(df['streams'], errors='coerce')

# Ordenar os dados por 'streams' em ordem decrescente e pegar os 5 primeiros
top_5_artists = df[['artist(s)_name', 'streams']].sort_values(by='streams', ascending=False).head(5)

# Ajustar o tamanho da figura para que os rótulos não fiquem sobrepostos
plt.figure(figsize=(10, 6))

# Criar o gráfico de barras com os 5 artistas principais (vertical)
plt.bar(top_5_artists['artist(s)_name'], top_5_artists['streams'])

# Adicionar rótulos e título
plt.ylabel('Streams')
plt.xlabel('Artistas')
plt.title('Top 5 Artistas por Streams')

# Rotacionar os rótulos dos artistas no eixo X para melhor visualização
plt.xticks(rotation=45, ha='right')  

# Adicionar valores sobre as barras para mostrar a quantidade de streams
for i, value in enumerate(top_5_artists['streams']):
    plt.text(i, value + 1000000, f'{int(value):,}', ha='center', va='bottom', fontsize=10)

# Exibir o gráfico
plt.tight_layout()  # Para garantir que tudo caiba dentro da área da figura
plt.show()


```
